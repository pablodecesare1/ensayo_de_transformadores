```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

## el poder de numpy es tremendo!
data=np.genfromtxt('z_trafo.csv',delimiter=',')
```

```python
t=data[2:,0]
V=data[2:,1]  # tension de linea
I=data[2:,2]/10   # corriente
plt.figure(dpi=100)
plt.plot(t,V,t,I*5000)


```

```python
p=V*I
pr=np.mean(10*I**2)
pactiva=np.mean(p)-pr
print("potencia en la R",pr)
print("Pact ",pactiva)
p_line=np.ones(len(p))*pactiva
plt.figure(dpi=100)
plt.plot(t,p,t,p_line)
```
